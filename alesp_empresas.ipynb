# Verifica se empresas estão nas despesas

```python
despesas = pd.read_csv("resultados/despesas_dep_est_sp_8_junho.csv", sep=',',encoding = 'utf-8', converters={'CNPJ': lambda x: str(x), 'cnpj_raiz_fornecedor': lambda x: str(x)}, decimal=',')
```

```python
despesas.rename(columns={'Valor':'valor_fornecedor'}, inplace=True)
```

```python
despesas_atuais = despesas[(despesas['Ano'] == 2015) | (despesas['Ano'] == 2016) | (despesas['Ano'] == 2017) | (despesas['Ano'] == 2018)]
```

```python
def limpeza3(linha):
    documento = linha["CNPJ"]
    if len(documento) > 11:
        return documento[:8]
    else:
        return "documento_cpf_"
```

```python
despesas_atuais['cnpj_raiz_fornecedor'] = despesas_atuais.apply(limpeza3, axis = 1)
```

```python
despesas_empresas_deputados = pd.merge(deputados_socios_empresas.drop_duplicates('cnpj_raiz_empresas'), despesas_atuais, left_on='cnpj_raiz_empresas', right_on='cnpj_raiz_fornecedor')
```

```python
despesas_empresas_deputados.info()
```

```python
despesas_empresas_deputados.to_csv('resultados/despesas_empresas_deputados.csv', index=False)
```

```python
despesas_empresas_deputados.reset_index()
```

```python
# Tabela dinâmica - despesas com empresas de deputado
```

```python
pd.pivot_table(despesas_empresas_deputados,index=["Ano","Deputado", "cnpj_raiz_fornecedor", "Fornecedor"], values=["valor_fornecedor"], aggfunc=np.sum)
```
