```python
import json
import xmltodict
import pandas as pd
import unidecode
from datetime import date
```

## Comissões

Fonte
https://www.al.sp.gov.br/dados-abertos/recurso/43

```python
with open("dados/processo_legislativo/comissoes_julho_2018.xml", 'r') as f:
    xmlString = f.read()
    
jsonString = json.dumps(xmltodict.parse(xmlString), indent=4)

with open("output.json", 'w') as f:
    f.write(jsonString)
```

```python
arquivo_json = open('output.json', 'r')
dados_json = json.load(arquivo_json)
```

```python
lista = dados_json['Comissoes']['Comissao']
```

```python
df_comissoes = pd.DataFrame(lista, columns = ['DataFimComissao', 'IdComissao', 'NomeComissao', 'SiglaComissao'])
```

```python
df_comissoes['DataFimComissao'] = pd.to_datetime(df_comissoes['DataFimComissao'])
```

```python
df_comissoes.info()
```

```python
df_comissoes.reset_index().head(10)
```

Seleciona apenas comissões permanentes ou que acabaram a partir de 2015

```python
mask = (df_comissoes['DataFimComissao'].dt.year >= 2015) | (df_comissoes['DataFimComissao'].isnull())

df_comissoes = df_comissoes[mask]
```

```python
df_comissoes.info()
```

```python
df_comissoes.reset_index().head(10)
```

## Membros de Comissões

Fonte: https://www.al.sp.gov.br/dados-abertos/dataset/8

```python
with open("dados/processo_legislativo/comissoes_membros_julho_2018.xml", 'r') as f:
    xmlString = f.read()
    
jsonString = json.dumps(xmltodict.parse(xmlString), indent=4)

with open("output.json", 'w') as f:
    f.write(jsonString)
```

```python
arquivo_json = open('output.json', 'r')
dados_json = json.load(arquivo_json)
```

```python
lista = dados_json['ComissoesMembros']['MembroComissao']
```

```python
df_membros = pd.DataFrame(lista, columns = ['DataInicio', 'Efetivo', 'IdComissao', 'IdMembro', 'IdPapel', 'NomeMembro', 'Papel', 'SiglaComissao' ])
```

```python
df_membros['DataInicio'] = pd.to_datetime(df_membros['DataInicio'])
```

```python
df_membros.info()
```

Seleciona apenas a partir de 2015

```python
mask = (df_membros['DataInicio'].dt.year >= 2015) | (df_membros['DataInicio'].isnull())

df_membros = df_membros[mask]
```

```python
df_membros.info()
```

```python
df_membros.reset_index().head(10)
```

```python
df_membros.to_excel('resultados/membrosdecomissoes.xlsx',sheet_name='Sheet1')
```

```python
df_membros.to_csv('resultados/membrosdecomissoes.csv', index=False)
```

Carrega nomes de deputados da atual legislatura

```python
nomes_deputados_atual_legislatura = pd.read_csv('resultados/deputados_selecionados_final.csv', converters={'sequencial': lambda x: str(x), 'cpf': lambda x: str(x), 'codigo_unico': lambda x: str(x) })
```

```python
nomes_deputados_atual_legislatura.info()
```

```python
df_membros_atuais = pd.merge(df_membros, nomes_deputados_atual_legislatura, left_on='IdMembro', right_on='codigo_unico')
```

```python
df_membros_atuais.info()
```

Verifica se há nomes não encontrados

```python
nomes_naoencontrados = pd.merge(df_membros, 
                        nomes_deputados_atual_legislatura, 
                        left_on='IdMembro', 
                        right_on='codigo_unico',
                        how='outer',
                        indicator=True)
```

```python
rdf = nomes_naoencontrados.query("_merge == 'right_only'").drop('_merge',axis=1)
```

```python
rdf[['nome_urna','situacao']]
```

## Presença nas Comissões

```python
with open("dados/processo_legislativo/comissoes_permanentes_presencas_julho_2018.xml", 'r') as f:
    xmlString = f.read()
    
jsonString = json.dumps(xmltodict.parse(xmlString), indent=4)

with open("output.json", 'w') as f:
    f.write(jsonString)
```

```python
arquivo_json = open('output.json', 'r')
dados_json = json.load(arquivo_json)
```

```python
#dados_json
```

```python
lista = dados_json['ComissoesReunioesPresencas']['ReuniaoComissaoPresenca']
```

```python
df_presencas = pd.DataFrame(lista, columns = ['DataReuniao', 'IdComissao', 'IdDeputado', 'IdPauta', 'IdReuniao', 'Deputado', 'SiglaComissao' ])
```

```python
df_presencas['DataReuniao'] = pd.to_datetime(df_presencas['DataReuniao'])
```

```python
df_presencas.info()
```

```python
df_presencas.reset_index().head()
```

Seleciona apenas a partir de 2015

```python
mask = (df_presencas['DataReuniao'].dt.year >= 2015) | (df_presencas['DataReuniao'].isnull())

df_presencas = df_presencas[mask]
```

```python
df_presencas.info()
```

Deputados mais membros de comissões

```python
df_comissao_membros = pd.merge(df_comissoes, df_membros_atuais, left_on='IdComissao', right_on='IdComissao')
```

```python
df_comissao_membros.groupby('NomeMembro')['IdComissao'].nunique().sort_values(ascending=False)
```

```python
df_comissao_membros.to_excel('resultados/deputados_membros_comissoes.xlsx',sheet_name='Sheet1')
```

Deputados mais presentes em comissões

```python
df_comissoes_presencas = pd.merge(df_presencas, df_comissoes, left_on='IdComissao', right_on='IdComissao')
```

```python
df_comissoes_presencas_selecao = pd.merge(df_comissoes_presencas , nomes_deputados_atual_legislatura, 
                                          left_on='IdDeputado', right_on='codigo_unico')
```

```python
df_comissoes_presencas_selecao.to_excel('resultados/comissoes_presencas_selecao.xlsx',sheet_name='Sheet1')
```
